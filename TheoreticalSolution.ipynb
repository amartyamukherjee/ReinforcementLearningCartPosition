This jupyter notebook uses the theoretical voltage function:
$$V(t)=K_p(r-x(t))$$

```
import matplotlib.pyplot as plt
import numpy as np
```

```
# Define the constants
alpha = -1
beta = 10
step_size = 0.2

time_limit = 20

A = np.array([[0,1],[0,alpha]])
B = np.array([[0],[beta]])

R = 10
```

# K = 0.2

```
# Run numerical simulations of the ODE using Euler's method
K = 0.2

x = np.array([[0],[0]])

time = 0

positions = []

while time < time_limit:
  v = K*(R-x[0,0])
  x = x + step_size*(A.dot(x)+B*v)
  time += step_size
  print("Voltage: ", v, "Position: ", x[0,0], "Velocity: ", x[1,0])

  positions.append(x[0,0])
```

```
plt.plot(positions)
plt.xlabel("Time step")
plt.ylabel("Distance from origin")
plt.savefig("Theoretical2.png")
plt.show()
```
